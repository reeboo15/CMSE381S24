```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import random
from sklearn.metrics import mean_squared_error
```

```python
random.seed(1234)

def true_model(x):
    return x**2 + np.random.normal(0, 1, len(x))

x_train = np.random.uniform(-2, 2, 400)
y_train = true_model(x_train)
```

```python
x_test = np.random.uniform(-2, 2, 400)
y_test = true_model(x_test)
```

```python
plt.scatter(x_train, y_train)
```

```python
plt.scatter(x_test, y_test)
```

```python
fit = np.polyfit(x_train, y_train,1)
print("beta_0:", fit[0], "beta_1:", fit[1])
```

```python
def fit_model(x):
    return fit[0]+fit[1]*x

y_train_fit = fit_model(x_train)
y_test_fit = fit_model(x_test)

mse_train = mean_squared_error(y_train, y_train_fit)
mse_test = mean_squared_error(y_test, y_test_fit)
print("Mean squared error for training dataset is", mse_train, ". Mean squared error for test dataset is", mse_test, ".")
```

```python
fit_list = []
deg_list = []
for a in range(1,21):
    fit_a = np.polyfit(x_train,y_train,a)
    polfit1 = np.polyval(fit_a, x_train)    
    mse_a1 = mean_squared_error(y_train, polfit1)
    polfit2 = np.polyval(fit_a, x_test)    
    mse_a2 = mean_squared_error(y_test, polfit2)
    fit_list.append([mse_a1,mse_a2])
    deg_list.append(a)
    print("For degree", a, ", the MSE for training dataset is", mse_a1, "The MSE for test dataset is", mse_a2)
    continue
```

```python
train_fit = []
test_fit = []
for i in fit_list:
    train_fit.append(i[0])
    test_fit.append(i[1])
    
plt.plot(deg_list, train_fit)
#this is for train dataset
```

```python
plt.plot(deg_list, test_fit)
#this is for test dataset
```

Yes, this matches my intuition because the MSE drops significantly when we match the degree of the model, but barely changes the more we increase the degree of the polyfit. In fact, the MSE increases when we overfit too much (at around 18 degree polynomial fit and higher).

```python
#starting new model, x^2 + 4 sin(10x) + n
random.seed(42)

def true_model2(x):
    return x**2 + 4*np.sin(10*x) + np.random.normal(0, 1, len(x))

x_train2 = np.random.uniform(-2, 2, 400)
y_train2 = true_model2(x_train2)
x_test2 = np.random.uniform(-2, 2, 400)
y_test2 = true_model2(x_test2)
plt.scatter(x_train2, y_train2)
```

```python
plt.scatter(x_test2, y_test2)
```

```python
fit2 = np.polyfit(x_train2, y_train2,1)
print("beta_0:", fit2[0], "beta_1:", fit2[1])
```

```python
def fit_model2(x):
    return fit2[0]+fit2[1]*x

y_train_fit2 = fit_model2(x_train2)
y_test_fit2 = fit_model2(x_test2)

mse_train2 = mean_squared_error(y_train2, y_train_fit2)
mse_test2 = mean_squared_error(y_test2, y_test_fit2)
print("Mean squared error for training dataset is", mse_train2, ". Mean squared error for test dataset is", mse_test2, ".")
```

```python
fit_list2 = []
deg_list2 = []
for a2 in range(1,21):
    fit_a2 = np.polyfit(x_train2,y_train2,a2)
    polfit12 = np.polyval(fit_a2, x_train2)    
    mse_a12 = mean_squared_error(y_train2, polfit12)
    polfit22 = np.polyval(fit_a2, x_test2)    
    mse_a22 = mean_squared_error(y_test2, polfit22)
    fit_list2.append([mse_a12,mse_a22])
    deg_list2.append(a2)
    print("For degree", a2, ", the MSE for training dataset is", mse_a12, "The MSE for test dataset is", mse_a22)
    continue
```

```python
train_fit2 = []
test_fit2 = []
for i2 in fit_list2:
    train_fit2.append(i2[0])
    test_fit2.append(i2[1])
    
plt.plot(deg_list2, train_fit2)
#this is for train dataset
```

```python
plt.plot(deg_list2, test_fit2)
#this is for test dataset
```

For this function, the MSE decreases at certain degrees (the plot looks like a staircase) and it does not seem very intuitive to me. The second model is more complex since it takes more time to generate a model that decreases MSE to a reasonable amount. The first model's MSE decreases by degree 2, but the second model's MSE decreases under 2 only after a model of degree 18 or 19.
