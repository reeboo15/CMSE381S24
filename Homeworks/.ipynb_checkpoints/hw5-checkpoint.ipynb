## 4.8.5
We now examine the differences between LDA and QDA. 

(a) If the Bayes decision boundary is linear, do we expect LDA or QDA to perform better on the training set? On the test set?

We expect QDA to be better than LDA on the training set. However since the boundary is linear, QDA is likely to overfit and so LDA will perform better than QDA here, on the test set. 

(b) If the Bayes decision boundary is non-linear, do we expect LDA or QDA to perform better on the training set? On the test set?

On the training set we would expect QDA to be better. Since the boundary is non-linear, QDA is also expected to be better on the test set. 

## 4.8.12

(a) Suppose that you wish to classify an observation $X \in R$ into apples and oranges. You fit a logistic regression model and find that 

$\hat{Pr}(Y = orange| X = x) =$
$\frac{exp(\hat{B_0} + \hat{B_1}x)}{1+exp(\hat{B_0} + \hat{B_1}x)}$

Your friend fits a logistic regression model to the same data using the softmax formulation in (4.13), and finds that

$\hat{Pr}(Y = orange| X = x) =$
$\frac{exp(\hat{a_{orange_0}} + \hat{a_{orange_1}}x)}{exp(\hat{a_{orange_0}} + \hat{a_{orange_1}}x)+exp(\hat{a_{apple_0}} + \hat{a_{apple_1}}x)}$

(a) What is the log odds of orange versus apple in your model?

To find the log odds we need to calculate log of $\frac{P}{1 - P}$. 

$P = \frac{exp(\hat{B_0} + \hat{B_1}x)}{1+exp(\hat{B_0} + \hat{B_1}x)}$

$1 - P = 1 - \frac{exp(\hat{B_0} + \hat{B_1}x)}{1+exp(\hat{B_0} + \hat{B_1}x)} = \frac{1}{1+exp(\hat{B_0} + \hat{B_1}x)}$

$\frac{P}{1-P} = exp(\hat{B_0} + \hat{B_1}x)$ 

$log(\frac{P}{1-P}) = \hat{B_0} + \hat{B_1}x$

(b) What is the log odds of orange versus apple in your friend’s model?

$\frac{P}{1-P} = \frac{exp(\hat{a_{orange_0}} + \hat{a_{orange_1}}x)}{exp(exp(\hat{a_{apple_0}} + \hat{a_{apple_1}}x)}$

$log (\frac{P}{1-P}) = \hat{a_{orange_0}} + \hat{a_{orange_1}}x - \hat{a_{apple_0}} - \hat{a_{apple_1}}x$

(c) Suppose that in your model, $\hat{B_0} = 2$ and $\hat{B_1} = -1$. What are the coefficient estimates in your friend’s model? Be as specific as possible.

$\hat{a_{orange_0}} - \hat{a_{apple_0}} = 2$

$\hat{a_{orange_1}} - \hat{a_{apple_1}} = -1$

(d) Now suppose that you and your friend fit the same two models on a different data set. This time, your friend gets the coefficient estimates $\hat{a_{orange_0}} = 1.2$, $\hat{a_{orange_1}} = -2$, $\hat{a_{apple_0}} = 3$, $\hat{a_{apple_1}} = 0.6$. What are the coefficient estimates in your model?

$\hat{B_0} = -1.8, \hat{B_1} = --2.6$

(e) Finally, suppose you apply both models from (d) to a data set with 2,000 test observations. What fraction of the time do you expect the predicted class labels from your model to agree with those from your friend’s model? Explain your answer.

We would expect them to agree 100% of the time since we are essentially using the same model.

(a)  Create a binary variable, mpg01, that contains a 1 if mpg contains a value above its median, and a 0 if mpg contains a value below its median. You can compute the median using the median() method of the data frame. Note you may find it helpful to add a column mpg01 to the data frame by assignment. Assuming you have stored the data frame as Auto, this can be done as follows: Auto['mpg01'] = mpg01.

```python
auto = pd.read_csv('Auto.csv')
auto = auto.replace('?', np.nan)
auto = auto.dropna()
auto.horsepower = auto.horsepower.astype('int')

median_mpg =  auto.iloc[:,0].median()

mpg01 = []

for i in range(len(auto)):
    if auto.iloc[i,0] > median_mpg:
        mpg01.append(1)
    else:
        mpg01.append(0)
        
auto['mpg01'] = mpg01
auto.head()
```

(c) Perform LDA, QDA, and logistic regression, on the training data in order to predict mpg01 from horsepower. What is the test error of the model obtained?

```python
X_train = auto_1.iloc[:, 3].values
X_train = np.reshape(X_train, (-1,1))

y_train = auto_1.iloc[:, 9].values
y_train = np.reshape(y_train, (-1,1))

X_test = auto_2.iloc[:, 3].values
X_test = np.reshape(X_test, (-1,1))

y_test = auto_2.iloc[:, 9].values
y_test = np.reshape(y_test, (-1,1))


lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

y_pred_lda = lda.predict(X_test)

accuracy_lda = accuracy_score(y_test, y_pred_lda)
print("Accuracy of the LDA model:", accuracy_lda)

test_error_lda = 1 - accuracy_lda
print("Test Error for LDA model:", test_error_lda)

qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)

y_pred_qda = qda.predict(X_test)

accuracy_qda = accuracy_score(y_test, y_pred_qda)
print("Accuracy of the QDA model:", accuracy_qda)

test_error_qda = 1 - accuracy_qda
print("Test Error for QDA model:", test_error_qda)

lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Accuracy of the Logistic Regeresion model:", accuracy_lr)

test_error_lr = 1 - accuracy_lr
print("Test Error for Logistic Regression model:", test_error_lr)
```

(d) Perform KNN, with various choices of K, use LOOCV to decide which value of K seems to perform the best.

```python
#used gpt to help write the code

X = auto.iloc[:, 3].values
X = np.reshape(X, (-1,1))
y = auto.iloc[:, 9].values
y = np.reshape(y, (-1,1))


k_vals = list(range(1,21))
mean_accuracy_scores = []

for k in k_vals:
    knn = KNeighborsClassifier(n_neighbors = k)
    accuracy_scores = cross_val_score(knn, X,y,cv = None)
    mean_accuracy = accuracy_scores.mean()
    mean_accuracy_scores.append(mean_accuracy)
    
best_k_index = np.argmax(mean_accuracy_scores)
best_k = k_vals[best_k_index]
best_accuracy = mean_accuracy_scores[best_k_index]

print("Best K value:", best_k)
print("Mean accuracy with best K value:", best_accuracy)
```
