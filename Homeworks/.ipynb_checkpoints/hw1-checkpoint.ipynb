## 2.4.1

For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size n is extremely large, and the number of predictors p is small.
Flexible statistical learning method will be expected to perform better. We have a larger sample size, which helps the model learn better. There are a small number of predictors which lessens the risk of overfitting. 

(b) The number of predictors p is extremely large, and the number of observations n is small.
Inflexible statistical learning method will be expected to perform better. A smaller sample size means there may be a large risk of bias if we use a flexible statistical learning method. The large number of predictors may cause a flexible statistical method to overfit. An inflexible statistical learning method will help avoid these since it could reach a better model faster.

(c) The relationship between the predictors and response is highly non-linear.
A flexible statistical learning method will be better since it can adapt itself to the non-linear data better than the inflexible method. Inflexible methods will try to warp the observations to its own idea of the model,  for example a best fit linear model. 


(d) The variance of the error terms, i.e. Ïƒ2 = Var("), is extremely high.
An inflexible statistical learning method will be better. A flexible method would try to incorporate the extra noise/ error terms into its model, and give an inaccurate model. 

## 2.4.3

We now revisit the bias-variance decomposition.

(a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.


```python
#this is taken from https://botlnec.github.io/islp/sols/chapter2/exercise3/

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 10.0, 0.02)

def squared_bias(x):
    return .002*(-x+10)**3
def variance(x):
    return .002*x**3 
def training_error(x):
    return 2.38936 - 0.825077*x + 0.176655*x**2 - 0.0182319*x**3 + 0.00067091*x**4
def test_error(x):
    return 3 - 0.6*x + .06*x**2
def bayes_error(x):
    return x + 1 - x

plt.xkcd()
#frame = plt.gca()
#frame.axes.xaxis.set_ticklabels([])
plt.figure(figsize=(10, 8))
plt.plot(x,squared_bias(x), label='squared bias')
plt.plot(x, variance(x), label='variance')
plt.plot(x, training_error(x), label='training error')
plt.plot(x, test_error(x), label='test error')
plt.plot(x, bayes_error(x), label='Bayes error')
plt.legend(loc='upper center')
plt.xlabel('model flexibility')
plt.show()

#arbitrary units
```

(b) Explain why each of the five curves has the shape displayed in part (a). typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves

Typical squared bias: As the model becomes more flexible, it can adapt more to the data and reduce the quantity of typical (squared) bias. 

Variance: The variance increases as the flexibility increases. This is because as the model becomes more flexible it captures more and more of the noise in the data, which increases variance. 

Training error: The training error decreases as flexibility increases. As the model becomes more flexible, it captures the data better (to the point of overfitting) and so the predictions will not be far from the observations. This decreases training error.

Test error: The test error has a parabolic shape with a minimum. This is found by adding the variance, Bayes error and squared bias. When either variance or squared bias is decreasing, the other increases significantly. Bayes error remains constant throughout. This results in the parabolic shape for the test error.

Bayes error: This is constant.

## 2.4.7(a), (c)

The table below provides a training data set containing six observations, three predictors, and one qualitative response variable. (coded into vectors)

```python
x_1 = np.array((0,2,0,0,-1,1))
x_2 = np.array((3,0,1,1,0,1))
x_3 = np.array((0,0,3,2,1,1))
y = ["Red", "Red","Red","Green","Green","Red"]
```

Suppose we wish to use this data set to make a prediction for Y when X1 = X2 = X3 = 0 using K-nearest neighbors. 

a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

```python
test_0 = [0,0,0,0,0,0]
dist1 = np.linalg.norm(x_1 - test_0)
dist2 = np.linalg.norm(x_2 - test_0)
dist3 = np.linalg.norm(x_3 - test_0)

print(dist1, dist2, dist3)
```

c) What is our prediction with K = 3? Why?

```python
#taken from https://www.w3schools.com/python/python_ml_knn.asp

from sklearn.neighbors import KNeighborsClassifier

data = list(zip(x_1,x_2,x_3))
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(data, y) 
```

```python
new_1 = 2
new_2 = 0
new_3 = 0
new_point = [(new_1,new_2,new_3)]

prediction = knn.predict(new_point)
prediction
```
